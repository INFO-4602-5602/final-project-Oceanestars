```python
import pandas as pd
import altair as alt
from vega_datasets import data
# Create a Dataframe from a CSV
dataCSV = pd.read_csv('asianPopulationbyState.csv')

# Drop rows with any empty cells
dataCSV.head()
```

```python
state_map = alt.topo_feature(data.us_10m.url, 'states')

click = alt.selection_multi(fields=['state'])

choropleth = (alt.Chart(state_map).mark_geoshape().transform_lookup(
    lookup='id',
    from_=alt.LookupData(dataCSV, 'id', ['Chinese', 'state']))
.encode(
    color='Chinese:Q',
    opacity=alt.condition(click, alt.value(1), alt.value(0.2)),
    tooltip=['state:N', 'Chinese:Q'])
.add_selection(click)
.project(type='albersUsa'))

bars = (
    alt.Chart(
        dataCSV.nlargest(15, 'Chinese'),
        title='Top 15 States by Chinese').mark_bar().encode(
    x='Chinese',
    opacity=alt.condition(click, alt.value(1), alt.value(0.2)),
    color='Chinese',
    y=alt.Y('state', sort='x'),
    tooltip = ['state:N', 'Chinese:Q'])
.add_selection(click))

# asianRestaurantsMap = alt.Chart(AsianRest).mark_circle().encode(
#     longitude='longitude:Q',
#     latitude='latitude:Q',
#     size=alt.value(20),
#     color=alt.value('#ff2400')
# #     tooltip=['name:N', 'address:N', 'city:N', 'state:N', 'categories:N']
# ).project(
#     "albersUsa"
# )

# restaurantMap = choropleth + asianRestaurantsMap
populationMap = choropleth | bars
populationMap
```

```python
populationMap.save("AsianClinicMap.html", embed_options={"render": "svg"})
```
