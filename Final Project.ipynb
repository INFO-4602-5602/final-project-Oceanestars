```python
import pandas as pd
import numpy as np
import altair as alt
from vega_datasets import data
dataStates = pd.read_csv("USCensus2019.csv")
censusStates = pd.read_csv("CensusState.csv")
censusStates.head()
#print(dataStates[['States', 'Asian Percent', 'Latitude', 'Longitude','Korean','Cambodian','Hmong']].head(10))
```

```python
alt.Chart(dataStates).mark_bar().encode(
    x = "States",
    y = "Asian Percent"
)
```

```python
#tried the interactive properties but it doesn't work with this map
states = alt.topo_feature(data.us_10m.url, feature='states')

# US states background
background = alt.Chart(states).mark_geoshape(
    fill='lightgray',
    stroke='white'
).properties(
    width=500,
    height=300
).project('albersUsa')

points = alt.Chart(dataStates).mark_circle().encode(
    longitude='Longitude:Q',
    latitude='Latitude:Q',
    size=alt.Size('Asian Percent:Q', title='% of Asian Population'),
    color=alt.value('steelblue'),
    tooltip=['States:N','Korean:Q','Cambodian:Q','Hmong:Q']
).properties(
    title='Asian population across the US'
)

AsianPop = background + points
AsianPop 
# AsianPop.save('AsianPop.html')
```

```python
#Attempting to have a dropdown of Asian subgroups and that when I click chinese
# all the states show up and just show the chinese column data so depending on the dropdown the 
# data for each state depends on the subgroup
dropdown = alt.binding_select(options=censusStates["Chinese","Indian"], name="Select a subgroup: ")
selection = alt.selection(type="single", fields=["column"], bind=dropdown, init={'column': "Chinese"})

alt.Chart(censusStates).transform_fold(
  ["Chinese","Indian"],
    as_=['column','value']
).transform_filter(selection
).mark_bar().encode(
    x = "States",
    y = "value:Q",
).add_selection(selection)
```
