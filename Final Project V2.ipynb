```python
# This is version two on narrowing down a graph for demographics
import pandas as pd
import numpy as np
import altair as alt
from vega_datasets import data
```

```python
state_pop = pd.read_csv("censusState.csv")


state_map = alt.topo_feature(data.us_10m.url, 'states')

click = alt.selection_multi(fields=['states'])

choropleth = (alt.Chart(state_map).mark_geoshape().transform_lookup(
    lookup='id',
    from_=alt.LookupData(state_pop, 'id', ['Asian Population', 'States']))
.encode(
    color='Asian Population:Q',
    opacity=alt.condition(click, alt.value(1), alt.value(0.2)),
    tooltip=['States:Q','Asian Population:Q'])
.add_selection(click)
.project(type='albersUsa'))

bars = (
    alt.Chart(
        state_pop.nlargest(15, 'Asian Population'),
        ).mark_bar().encode(
    x='Asian Population',
    opacity=alt.condition(click, alt.value(1), alt.value(0.2)),
    color='Asian Population',
    y=alt.Y('States', sort='x'))
.add_selection(click))

FinalMap = choropleth | bars
FinalMap
# FinalMap.save("Asian.html")
```
